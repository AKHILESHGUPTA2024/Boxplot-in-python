```python
import matplotlib.pylab as plt
import numpy as np
```

```python
data = np.random.normal(100, 20, 200)
plt.boxplot(data)
```

```python
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

labels = ['peachpuff', 'orange', 'tomato','black']
colors = ['darkblue', 'orange', 'red','yellow']


a=plt.boxplot(data,notch=True,patch_artist=True,sym='o',vert=False,whis=True,positions=None,widths=.4,bootstrap=1000,showbox=True,showcaps=True,
            showfliers=True,showmeans=False, meanline=False,boxprops=dict(color="black"),whiskerprops=dict(color="red"),capprops=dict(color="green"),
            medianprops=dict(color="orange"),tick_labels=labels)
for patch, color in zip(a['boxes'], colors):
    patch.set_facecolor(color)
    
```

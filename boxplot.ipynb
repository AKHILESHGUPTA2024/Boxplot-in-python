```python
Jan21= np.random.normal(100,90,200)
Feb21= np.random.normal(100,80,200)
Mar21= np.random.normal(100,40,120)
Apr21= np.random.normal(100,90,180)
May21= np.random.normal(100,90,190)
Jun21= np.random.normal(90,90,200)
Jul21= np.random.normal(80,100,200)
Aug21= np.random.normal(100,90,200)
Sep21= np.random.normal(75,120,200)
Oct21= np.random.normal(100,78,170)
Nov21= np.random.normal(89,90,140)
Dec21= np.random.normal(100,30,300)

fig = plt.figure(figsize=(14,5))

Year2021 = [Jan21,Feb21,Mar21,Apr21,May21,Jun21,Jul21,Aug21,Sep21,Oct21,Nov21,Dec21]
label = ['Jan21','Feb21','Mar21','Apr21','May21','Jun21','Jul21','Aug21','Sep21','Oct21','Nov21','Dec21']

# colors = ['red','black','blue','grey','purple','yellow','pink','magenta','darkblue','skyblue','orange','blue']
# colors = ['#ff7f7f','#ff0000','#674867','#ffbf00','#ffd700','#3b3c36','#ddd06a','#f4a460','#DFFF00','#D4af37','#808000','#00ffbf']
colors = plt.get_cmap('Greys')(np.linspace(0.2,0.9,len(Year2021)))
x = plt.boxplot(Year2021,notch=True,labels=label,patch_artist=True)

for patch, color in zip(x['boxes'],colors):
    patch.set_facecolor(color)

for median in x['medians']:
    median.set(color='red',linewidth=2)
for flier in x['fliers']:
    flier.set(color='white',linewidth=2,mfc='r',mec='white',ms=10)
```
